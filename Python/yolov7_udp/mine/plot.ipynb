```python
import os
import matplotlib.pyplot as plt

# Path to YOLO annotations
yolo_dir = "/media/holidayj/Documents/data/datasets_for_object_detection/coco/images/train2017_yolo"

# Collect all widths and heights
widths = []
heights = []

# Iterate over all .txt files
for filename in os.listdir(yolo_dir):
    if not filename.endswith(".txt"):
        continue

    with open(os.path.join(yolo_dir, filename), 'r') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) != 5:
                continue
            w = float(parts[3])  # width
            h = float(parts[4])  # height
            widths.append(w)
            heights.append(h)

# Plotting
plt.figure(figsize=(12, 5))

# Width histogram
plt.subplot(1, 2, 1)
plt.hist(widths, bins=50, color='skyblue', edgecolor='black')
plt.title('Width Distribution')
plt.xlabel('Width (normalized)')
plt.ylabel('Frequency')

# Height histogram
plt.subplot(1, 2, 2)
plt.hist(heights, bins=50, color='salmon', edgecolor='black')
plt.title('Height Distribution')
plt.xlabel('Height (normalized)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import os
import matplotlib.pyplot as plt

# Configure global font to Times New Roman and increase font sizes
plt.rcParams.update({
    "font.family": "Times New Roman",
    "font.size": 24,
    "axes.titlesize": 24,
    "axes.labelsize": 24,
    "xtick.labelsize": 22,
    "ytick.labelsize": 22,
    "legend.fontsize": 22
})

# Path to YOLO annotations
yolo_dir = "/media/holidayj/Documents/data/datasets_for_object_detection/coco/images/train2017_yolo"

# Collect all widths and heights
widths = []
heights = []

# Iterate over all .txt files
for filename in os.listdir(yolo_dir):
    if not filename.endswith(".txt"):
        continue

    with open(os.path.join(yolo_dir, filename), 'r') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) != 5:
                continue
            w = float(parts[3])  # width
            h = float(parts[4])  # height
            widths.append(w)
            heights.append(h)

# Plot both histograms in one figure with log scale
plt.figure(figsize=(10, 6))

bins = 100
plt.hist(widths,  bins=bins, alpha=0.8, label='Width', color='skyblue', edgecolor='black', log=True)
plt.hist(heights, bins=bins, alpha=0.5, label='Height', color='salmon', edgecolor='black', log=True)

# plt.title('Normalized Width and Height Distribution (Log Scale)')
plt.xlabel('Normalized Size')
plt.ylabel('Frequency (Log Scale)')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```

```python
import os
import matplotlib.pyplot as plt

# Path to YOLO annotations
yolo_dir = "/media/holidayj/Documents/data/datasets_for_object_detection/coco/images/train2017_yolo"
# yolo_dir = "/home/holidayj/Desktop/yolov7_udp/data/train"

# Collect all widths, heights, and areas
widths = []
heights = []
areas = []

# Iterate over all .txt files
for filename in os.listdir(yolo_dir):
    if not filename.endswith(".txt"):
        continue

    with open(os.path.join(yolo_dir, filename), 'r') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) != 5:
                continue
            w = float(parts[3])  # width
            h = float(parts[4])  # height
            widths.append(w)
            heights.append(h)
            areas.append(w * h)

# Plot all histograms in one figure
plt.figure(figsize=(10, 6))
bins       = 100
bins_areas = 100
plt.hist(heights, bins=bins, alpha=0.5, label='Height', color='salmon', edgecolor='black', log=True)
plt.hist(widths, bins=bins, alpha=0.5, label='Width', color='skyblue', edgecolor='black', log=True)
plt.hist(areas, bins=bins_areas, alpha=0.5, label='Area', color='mediumseagreen', edgecolor='black', log=True)

# plt.title('Normalized Width, Height, and Area Distributions (Log Scale)')
plt.xlabel('Normalized Width, Height, and Area in the MS COCO dataset')
plt.ylabel('Frequency (log scale)')
plt.xlim(0,1)

handles, labels = plt.gca().get_legend_handles_labels() 
order = [1, 0,2 ] 
plt.legend([handles[i] for i in order], [labels[i] for i in order]) 

# plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```
